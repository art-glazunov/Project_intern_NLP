```python
import os
import gc

import numpy as np
import pandas as pd


from matplotlib import pyplot as plt
import seaborn as sns 

from tqdm import tqdm_notebook

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split



import my_preprocessing #my little prep module in the project folder (requires num2words, pymorphy2, natasha, pymystem3==0.1.10)
```

```python
def preprocessing(text):
  #Prepares data using my module
                 
  text_res = my_preprocessing.text_early_preproc(text,del_html = True,del_punct_sp_chars=True,
                 del_underscore=False, del_digits=False) #delete punctuation and artifacts

  text_res = my_preprocessing.lemmatize_lower_case2(text_res)
 
  text_res = my_preprocessing.delete_stop_words(text_res)

  return text_res
```

Let's preprocess our data, previously deleting NANs

```python
folder_to_save = 'texts_after_preproc'
```

```python
#Do not launch in vain, a long process (6 hours)

prepared_texts = []
counter = 0
for ind,text in enumerate(tqdm_notebook(data[['text']].dropna().text.values)):

  prepared_texts.append(preprocessing(text))

  if (ind+1) % 1e5 == 0:
    #save parts
    counter +=1

    text_saved = pd.DataFrame(prepared_texts)
    text_saved.to_csv(os.path.join(folder_to_save,f'preproc_texts{counter}.csv'))
  

prepared_texts = pd.DataFrame(prepared_texts)
prepared_texts.to_csv(os.path.join(folder_to_save,f'preproc_texts.csv')) #Preprocessed texts
```
